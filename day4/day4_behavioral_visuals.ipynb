```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

it changes the plots styles to a more presentable way

```python
data= pd.read_csv("https://raw.githubusercontent.com/dasarpai/DAI-Datasets/main/WA_Fn-UseC_-HR-Employee-Attrition.csv")
data.head(5)
```

```python
#Plot Age vs YearsAtCompany using scatter
X, Y= data['Age'], data['YearsAtCompany']

plot_1= plt.scatter(X,Y ,c='red', alpha=0.6, edgecolors='black', )
plt.show
plt.xlabel('Age')
plt.ylabel('Years At Company')
plt.title('Age vs YearsAtCompany')
plt.show()
```

the above scatter plot shows years of an employee according to the age, from the plot we got an insight that younger age group have less years at company while the plottings increase with age the years at company also increses, though at older age group employees are less. scatter plots are use when we have to get relation between two numerical (continuous) variables

```python
#Plot average MonthlyIncome by Department using bar chart
X1 = data.groupby('Department')['MonthlyIncome'].mean()

X2= data['Department']
plot=plt.bar(X1.index,X1.values)
plot
plt.ylabel('avg mothly income')
plt.xlabel('Department')
plt.title('avg salary per department')
pat= ['*', 'o', '/']
for bar in plot:
    bar.set_hatch(pat.pop(0))
    
```

Always group and aggregate your data before using bar charts when working with columns like Department, JobRole, etc.
the plot above showcases the avg salary of the employees based on their education field where we gwt an idea that there are largest no of employees in sales department. bar plots are used to compare categories, where xaxis is categories and y axis is th enumeric data.

```python
#Plot histogram of JobSatisfaction
Xu= data['JobSatisfaction']
plt.hist(Xu,bins=[0.5,1.5,2.5,3.5,4.5], edgecolor='black')
plt.xlabel('distribution of job satisfaction')
plt.ylabel('employees')
plt.title('distribution of job satisfaction of the employees ')
plt.xticks(ticks=[1,2,3,4], labels=['low','medium-low', 'medium-high','high' ])
```

from above histogram we got a distribution of job satifaction at the organization among employees, it shows insight that 400+ people chooses high job satisfaction. histograms are used to see how data is distributed (frequency), and when a single column is specified to anlyze

```python
#Find correlations between:
#YearsAtCompany, JobSatisfaction, MonthlyIncome, Age
col= ['YearsAtCompany', 'JobSatisfaction', 'MonthlyIncome', 'Age']
correlation= data[col].corr(numeric_only= True)
correlation
```

```python
sns.heatmap(correlation, annot= True ,cmap='coolwarm', linecolor= 'white',linewidth=0.5, fmt ='.2f')
```

```python
heatmap is used  to show relationships or patterns in a matrix,  analyzing a correlation matrix or any tabular comparison,quick visual overview of many variables
```

```python
#boxplot() comparing JobRole vs MonthlyIncome.
sns.boxplot(data= data, x='JobRole' , y= 'MonthlyIncome')
```

To summarize distributions quickly ,To compare distributions between categories, To spot outliers and skewness in data
