```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
data= pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
```

```python
data
```

```python
X= data[['Hours']] # indxdependent variable in x variable (2d)
Y=data['Scores'] #dependant variable in y axis
```

```python
model= LinearRegression()
```

```python
model.fit(X,Y)
```

```python
print(model.intercept_)
```

```python
y_predict= model.predict(X) #predict by independent variables
print(model.predict([[3.4]]))
```

```python
plt.scatter(X,Y, label= 'actual')
plt.plot(X,y_predict, 'r', label= 'predicted')
plt.legend()
plt.xlabel('Hours')
plt.ylabel('Scores')
```

```python
print(mean_squared_error(Y, y_predict))
```

```python
print(r2_score(Y,y_predict))
```
